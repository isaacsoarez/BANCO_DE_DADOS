```python
import pyodbc
import pandas as pd

conexao = ("Driver={SQLite3 ODBC Driver};"
        "Server=localhost;"
        "Database=chinook.db")

conexao2 = pyodbc.connect(conexao)
print("Conexão Bem Sucedida")


cursor = conexao2.cursor()
```

```python
cursor.execute("SELECT * FROM customers")

values = cursor.fetchall()
print(values)

descricao = [column[0] for column in cursor.description]
print(descricao)
```

```python

tabela_customers = pd.DataFrame.from_records(values, columns=descricao)
print(tabela_customers)
```

```python
# Verifique se a tabela "customers" existe no banco de dados "chinook.db"
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='customers';")
tabela_existe = cursor.fetchone()

if tabela_existe:
    # A tabela existe, você pode prosseguir com sua consulta
    cursor.execute("SELECT * FROM customers")
    valores = cursor.fetchall()

    # Apresente as informações
    for valor in valores:
        print(valor)
else:
    print("A tabela 'customers' não existe no banco de dados.")
```
